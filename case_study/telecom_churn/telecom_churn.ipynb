### Since there are no columns now with more than 10% missing values we can handle these data

#### Check if there are columns with one unique value since it won't be beneficial our analysis

```python
telecom_df.nunique()
```

### Since there are few columns which have one unique values, lets drop it

```python
# lets find the columns that have only 1 unique value and drop such columns
unique_1_col=[]
for i in telecom_df.columns:
    if telecom_df[i].nunique() == 1:
        unique_1_col.append(i)
    else:
        pass

telecom_df.drop(unique_1_col, axis=1, inplace = True)
print("Columns dropped are -> \n", unique_1_col)
```

```python
# re-analyzing the unique columns
telecom_df.nunique()
```

```python
# check shape of dataset again
telecom_df.shape
```

#### Handling Missing Values

```python
# capturing the telecom data missing values to do further analysis
telecom_values_missing = telecom_df.isnull().mean()*100
telecom_values_missing

# filtering only columns which have missing values more than 0
telecom_values_missing = telecom_values_missing[telecom_values_missing > 0]
telecom_values_missing
```

#### As we see date type fields like date_of_last_rech_* is of type object, lets convert it to date type

```python
# selecting all the columns with datetime format
date_col= telecom_df.select_dtypes(include=['object'])
print("Date columns are -> \n",date_col.columns)

# Converting the selected columns to datetime format
for i in date_col.columns:
    telecom_df[i] = pd.to_datetime(telecom_df[i])
```
